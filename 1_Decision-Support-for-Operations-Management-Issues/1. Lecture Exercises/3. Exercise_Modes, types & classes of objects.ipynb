```r
x <- 1:5
y <- 6:10 
#print(x*y)
temp <- integer()

outer_new <-function(){
    for (i in x){
    for(j in y){
        var <- c(i*j)
        temp <- append(temp,i*j)
        
    }
 }
  return(temp)  
}

```

traceCom <-function(){
    while(iter <= nrow(sqMatrix)){
      while(iter2 <= ncol(sqMatrix)){
          if(iter == iter2){
              initi=initi + sqMatrix[iter,iter2]
          }
             
          iter2 <- iter + 1
      }  
        iter <- iter + 1
    }
    return(initi)
}
